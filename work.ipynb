## Date Fix

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.drop(columns=['timestamp'], inplace=True)
```

```python
df = df.sort_values(['crypto_name', 'date'])
df = df.reset_index(drop=True)
```

```python
print("date min:", df['date'].min())
print("date max:", df['date'].max())
print("columns:", df.columns.tolist())
```

```python
print("total cryptocurrencies:", df['crypto_name'].nunique())
print("\nsaari cryptos:")
print(df['crypto_name'].unique())
```

```python
## Numbers summary
print(df[['open', 'high', 'low', 'close', 'volume', 'marketCap']].describe())
```

```python
## Only for bitcoin Data
bitcoin = df[df['crypto_name'] == 'Bitcoin'].copy()
bitcoin = bitcoin.reset_index(drop=True)

print("bitcoin rows:", len(bitcoin))
print(bitcoin.head())
```

```python
## Remove unnamed column
bitcoin.drop(columns=['Unnamed: 0'], inplace=True)

print("columns:", bitcoin.columns.tolist())
print("shape", bitcoin.shape)
```

```python
## bitcoin_clean.csv Save
bitcoin.to_csv('bitcoin_clean.csv', index=True)
print("file saved!")
```
